# Consumption of Fruits and Vegetables

```python
dataFruitVeg = pd.read_csv("daily_consumption_fruit_vegetables_eu.csv")
```

```python
dataFruitVeg.head(10)
```

```python
#check if ireland is included
dataFruitVeg['country'].value_counts()
```

```python
#isolate ireland 
IEfruit = dataFruitVeg.loc[dataFruitVeg['country'] == ('IE')]
IEfruit.head()
```

```python
#check the amount of entries
IEfruit['country'].value_counts()
```

```python
IEfruit.plot(kind = 'bar', x='time', figsize=(15,10))
```

# Fish and Seafood

```python
dataFish = pd.read_csv("fish-and-seafood-consumption-per-capita.csv")
```

```python
dataFish.head(10)
```

```python
dataFishPie = dataFish.groupby('Entity')['Year'].sum()
```

```python
dataFishPie.plot.pie(figsize=(8,8), radius=1.2, autopct='%1.1f%%')
```
